# Bài tập Lý thuyết

## 1. State the applications of image inpainting.

Image inpainting là hành động khôi phục các vùng bị thiếu hoặc bị hỏng của hình ảnh bằng cách sử dụng thông tin từ các pixel xung quanh. Image inpainting có nhiều ứng dụng thực tế, chẳng hạn như khôi phục ảnh cũ, loại bỏ các đối tượng không mong muốn, lấp đầy các điểm che khuất và tạo bố cục hình ảnh chân thực. Các mô hình Machine Learning có thể học cách thực hiện việc vẽ hình ảnh bằng cách đào tạo trên bộ dữ liệu hình ảnh lớn và học các mẫu và cấu trúc của hình ảnh tự nhiên.

image.png

Image inpainting, hay còn gọi là kỹ thuật lấp đầy khoảng trống trong hình ảnh, là một công cụ mạnh mẽ với nhiều ứng dụng thiết thực trong đời sống hàng ngày. Kỹ thuật này có thể được sử dụng để loại bỏ các đối tượng không mong muốn khỏi hình ảnh, giúp cải thiện tính thẩm mỹ hoặc đảm bảo sự riêng tư trong các bức ảnh gia đình hoặc sự kiện. Ngoài ra, nó còn được áp dụng để khôi phục các bức ảnh cũ hoặc bị hỏng, giúp nâng cao chất lượng hoặc độ rõ nét của những kỷ niệm quý giá từ quá khứ.

Image inpainting còn được sử dụng để lấp đầy các khoảng trống hoặc lỗ hổng trong hình ảnh nhằm hoàn thiện thông tin còn thiếu. Điều này rất hữu ích trong trường hợp các tài liệu hoặc bức tranh cổ bị mất một phần nội dung. Hơn nữa, kỹ thuật này còn được tận dụng để tạo ra các hiệu ứng nghệ thuật đặc biệt, mang lại sự sáng tạo và bầu không khí mới mẻ cho các tác phẩm hình ảnh. Ngoài ra, nó cũng đóng vai trò quan trọng trong việc chỉnh sửa hình ảnh liên quan đến vấn đề bản quyền, chẳng hạn như xóa bỏ logo hoặc watermark trước khi tái sử dụng.

Dưới đây là hai ví dụ cụ thể trong đời sống hàng ngày để minh họa khả năng ứng dụng của image inpainting:

1. Khôi phục tranh tường cổ: Một bức tranh tường tại một ngôi đền hoặc nhà thờ bị hư hỏng do thời gian hoặc thiên tai có thể được phục hồi bằng cách sử dụng kỹ thuật inpainting. Các nhà phục chế nghệ thuật có thể lấp đầy các phần bị bong tróc hoặc mất đi của bức tranh, giữ nguyên được tính nguyên bản và giá trị lịch sử của nó. Nhờ đó, di sản văn hóa được bảo tồn, và người xem có thể chiêm ngưỡng trọn vẹn vẻ đẹp của tác phẩm.

2. Chỉnh sửa ảnh cưới: Trong một bức ảnh cưới, nếu có một vị khách không mời đứng lọt vào khung hình hoặc một vật thể không mong muốn, như một thùng rác, xuất hiện trong nền, kỹ thuật inpainting có thể được sử dụng để loại bỏ những yếu tố này. Kết quả là cặp đôi sẽ có một bức ảnh hoàn hảo, không bị ảnh hưởng bởi những chi tiết phá hỏng tổng thể.

Với những ứng dụng đa dạng như vậy, image inpainting đã trở thành một công cụ không thể thiếu, không chỉ trong lĩnh vực công nghệ mà còn trong các hoạt động thường nhật và nghệ thuật


## 2. State the applications of image inpainting in AI.

Trong lĩnh vực trí tuệ nhân tạo (AI), image inpainting không chỉ là một công cụ chỉnh sửa hình ảnh mạnh mẽ mà còn đóng vai trò quan trọng trong nhiều ứng dụng và nghiên cứu tiên tiến. Kỹ thuật này tận dụng deep learning như Generative Adversarial Networks (GANs) hoặc Autoencoders để tái tạo các phần hình ảnh bị thiếu, mang lại độ chính xác và tự nhiên cao

1. Trong các ứng dụng phân tích hình ảnh y tế, image inpainting có thể được sử dụng để loại bỏ các đối tượng không cần thiết hoặc các nhiễu ảnh như thiết bị y tế hoặc các vết mờ, từ đó cải thiện độ chính xác khi phân tích. Ví dụ, trong chụp MRI hoặc X-ray, kỹ thuật này có thể làm sạch ảnh để tập trung vào khu vực quan trọng cần được chẩn đoán.

2. Trong hệ thống giám sát, các camera an ninh có thể bị che khuất hoặc mất dữ liệu do thời tiết xấu hoặc nhiễu sóng. Image inpainting trong AI giúp tái tạo các phần hình ảnh bị mất, hỗ trợ việc nhận diện các sự kiện quan trọng hoặc đối tượng bị khuất trong khung hình.

3. Trong VR/AR, image inpainting giúp lấp đầy các khoảng trống hoặc chi tiết bị thiếu trong không gian ảo, mang lại trải nghiệm mượt mà và chân thực hơn. Ví dụ, khi tạo dựng một không gian 3D từ các bức ảnh chụp, AI có thể sử dụng kỹ thuật này để tái tạo những phần không được chụp rõ ràng.

4. Trong ngành sản xuất nội dung, image inpainting được sử dụng để tạo ra những bức ảnh hoàn hảo mà không cần phải chụp lại. Ví dụ, khi sản xuất phim hoặc video, AI có thể xóa đi các chi tiết thừa, như các thiết bị quay phim vô tình lọt vào khung hình, hoặc thêm vào các chi tiết hoàn toàn mới.

5. Trong các hệ thống nhận diện khuôn mặt, nếu một phần khuôn mặt bị che khuất (bởi khẩu trang, kính râm, hoặc các vật cản khác), image inpainting có thể tái tạo các phần khuôn mặt này, giúp cải thiện độ chính xác của hệ thống.

6. Khi làm việc với dữ liệu hình ảnh bị thiếu hoặc không đầy đủ, image inpainting có thể được sử dụng để tái tạo các phần bị thiếu, tạo ra một tập dữ liệu chất lượng cao hơn để huấn luyện các mô hình AI. Điều này đặc biệt hữu ích trong các dự án mà việc thu thập dữ liệu đầy đủ gặp nhiều khó khăn.

7. AI với image inpainting có thể tạo ra những tác phẩm nghệ thuật mới lạ bằng cách chỉnh sửa hoặc tái tạo các phần còn thiếu của các bức tranh nổi tiếng. Điều này không chỉ mở ra khả năng sáng tạo không giới hạn mà còn hỗ trợ việc giảng dạy và nghiên cứu nghệ thuật.

Nhờ những ứng dụng đa dạng này, image inpainting trong AI đã chứng minh tiềm năng lớn trong việc cải thiện chất lượng hình ảnh, tối ưu hóa các hệ thống thông minh và mở ra những hướng đi mới trong sáng tạo công nghệ.

## 3. Compare the advantages and disadvantages of image inpainting, image denoising, and image filtering.

**Image inpainting**

1. Ưu điểm:Kỹ thuật Image Inpainting mang lại nhiều ưu điểm vượt trội, góp phần giải quyết các vấn đề phức tạp trong xử lý hình ảnh. Đầu tiên, inpainting giúp khôi phục các vùng bị thiếu hoặc hỏng trong hình ảnh, chẳng hạn như các vết xước, lỗ hổng, hoặc vật cản, mang lại tính thẩm mỹ cao và cải thiện chất lượng hình ảnh. Kỹ thuật này đặc biệt hữu ích trong việc phục chế ảnh cũ, loại bỏ các đối tượng không mong muốn hoặc tái hiện các chi tiết bị mất trong quá trình chụp. Thứ hai, với sự hỗ trợ từ các mô hình học sâu như GANs hoặc Autoencoders, image inpainting có thể tái tạo các vùng ảnh với độ tự nhiên và chính xác cao, đảm bảo rằng kết quả khớp với ngữ cảnh và cấu trúc tổng thể của hình ảnh. Ngoài ra, inpainting còn ứng dụng trong sáng tạo nghệ thuật, giúp tạo ra các hiệu ứng độc đáo hoặc chỉnh sửa hình ảnh theo cách không giới hạn, mở ra tiềm năng lớn trong các lĩnh vực như thiết kế đồ họa, phim ảnh, và thực tế ảo. Cuối cùng, inpainting không chỉ cải thiện trải nghiệm người dùng mà còn hỗ trợ các ứng dụng chuyên môn như y tế, an ninh, và dữ liệu không gian, giúp tái hiện thông tin quan trọng một cách hiệu quả và chính xác.

2. Kỹ thuật Image Inpainting mặc dù mang lại nhiều tiềm năng nhưng vẫn tồn tại một số hạn chế đáng kể. Đầu tiên, việc thiếu các bộ dữ liệu lớn và đa dạng để huấn luyện và đánh giá mô hình khiến hiệu quả của kỹ thuật này bị giới hạn, đặc biệt khi các bộ dữ liệu hiện tại thường mang tính đồng nhất hoặc nhân tạo, không phản ánh đúng tính phức tạp của hình ảnh thực tế. Thứ hai, đảm bảo tính nhất quán về cấu trúc và ngữ nghĩa của vùng được khôi phục với toàn bộ hình ảnh là một thách thức, đặc biệt khi xử lý các vùng khuyết lớn hoặc phức tạp, có thể dẫn đến kết quả không hòa hợp với bối cảnh chung. Ngoài ra, kỹ thuật này còn thiếu khả năng kiểm soát và phản hồi từ người dùng, khiến người dùng khó can thiệp hoặc tùy chỉnh kết quả theo ý muốn, trong khi các mô hình học sâu thường hoạt động như “hộp đen” và thiếu tính minh bạch. Cuối cùng, hiệu suất tính toán và khả năng mở rộng cũng là vấn đề lớn khi một số mô hình đòi hỏi tài nguyên tính toán cao, thời gian xử lý dài, và không dễ dàng thích nghi với các hình ảnh có độ phân giải cao hoặc ứng dụng trong các miền khác nhau mà không cần tái huấn luyện.

**Image Denoising**

1. Ưu điểm:Image denoising mang lại nhiều ưu điểm quan trọng, đặc biệt trong xử lý tài liệu và các ứng dụng khác. Trước hết, nó cải thiện chất lượng hình ảnh bằng cách loại bỏ các loại nhiễu như hạt, đốm hoặc biến dạng màu sắc, giúp hình ảnh trở nên rõ nét và dễ sử dụng hơn. Điều này cực kỳ hữu ích trong các ứng dụng như chụp ảnh tài liệu, phục hồi ảnh kỹ thuật số, hoặc xử lý ảnh y tế. Ngoài ra, image denoising đóng vai trò then chốt trong việc nâng cao hiệu quả của công nghệ nhận diện ký tự quang học (OCR). Việc loại bỏ nhiễu và tăng độ rõ nét của văn bản trên hình ảnh giúp OCR nhận diện chính xác hơn, giảm thiểu lỗi và cải thiện hiệu suất xử lý dữ liệu. Hơn nữa, nó còn hỗ trợ các thuật toán phân đoạn, phân loại tài liệu, và các kỹ thuật tăng cường hình ảnh khác, tạo điều kiện thuận lợi cho việc xử lý và lưu trữ tài liệu trong nhiều lĩnh vực như doanh nghiệp, y tế, và giáo dục.

2. Nhược điểm: Image denoising dù mang lại nhiều lợi ích nhưng cũng tồn tại một số nhược điểm và hạn chế cần lưu ý. Trước tiên, quá trình loại bỏ nhiễu đôi khi có thể làm mất đi các chi tiết quan trọng trong hình ảnh, đặc biệt là khi xử lý các vùng có độ tương phản thấp hoặc chi tiết nhỏ. Điều này có thể dẫn đến việc giảm chất lượng hình ảnh hoặc làm sai lệch thông tin cần thiết. Thứ hai, các kỹ thuật denoising phức tạp thường đòi hỏi tài nguyên tính toán cao, dẫn đến thời gian xử lý lâu và tiêu tốn nhiều năng lượng, điều này gây khó khăn trong các ứng dụng thời gian thực hoặc trên các thiết bị có cấu hình thấp. Ngoài ra, hiệu quả của denoising phụ thuộc rất nhiều vào việc hiểu và mô hình hóa chính xác loại nhiễu trong ảnh; nếu loại nhiễu không được xác định đúng, các phương pháp loại bỏ có thể không đạt kết quả mong muốn. Cuối cùng, việc áp dụng các kỹ thuật tiên tiến như học sâu (deep learning) đòi hỏi lượng lớn dữ liệu huấn luyện, trong khi dữ liệu đủ đa dạng và chất lượng cao không phải lúc nào cũng sẵn có. Những hạn chế này đặt ra thách thức lớn trong việc tối ưu hóa hiệu suất và ứng dụng của image denoising.

**Image Filtering**

1. Ưu điểm: Một trong những lợi ích của việc áp dụng Image Filtering là khả năng làm nổi bật màu sắc và độ tương phản, giúp bức ảnh trở nên sống động và hấp dẫn hơn. Bộ lọc có thể tăng cường các chi tiết màu sắc trong ảnh, khiến chúng trở nên sáng hơn và bắt mắt hơn, như trong trường hợp ảnh chụp hoàng hôn. Bên cạnh đó, bộ lọc cũng có thể tạo ra hiệu ứng sắc nét cho ảnh, giúp bức ảnh trở nên độc đáo và dễ nhận diện. Ngoài ra, bộ lọc còn giúp tạo ra cảm giác chuyển động trong những bức ảnh tĩnh, tăng thêm sự sinh động và hấp dẫn cho bức ảnh.

2. Nhược điểm: Mặc dù bộ lọc ảnh có thể tạo ra những hiệu ứng ấn tượng, nhưng đôi khi chúng cũng mang lại những tác động tiêu cực. Một số bộ lọc có thể làm mất đi thông tin quan trọng trong ảnh, khiến bức ảnh trở nên quá mịn màng hoặc nhân tạo. Hơn nữa, việc sử dụng bộ lọc quá mức có thể làm giảm độ sắc nét của các cạnh, gây ra hiện tượng mờ hoặc tạo ra các tác dụng phụ không mong muốn như vầng sáng hay nhiễu. Đặc biệt, một số bộ lọc có thể thúc đẩy các chuẩn mực vẻ đẹp không thực tế, làm tổn hại đến tâm lý và lòng tự trọng của người dùng.










## 4. State the applications of RGB and HSV color channels.

RGB color channels là một mô hình màu bổ sung trong đó các màu cơ bản đỏ, lục và xanh dương của ánh sáng được thêm vào với nhau theo nhiều cách khác nhau để tái tạo một mảng màu rộng. Tên của mẫu này bắt nguồn từ tên viết tắt của ba màu cơ bản phụ là đỏ, lục và xanh dương.
Ứng dụng của các kênh màu RGB rất đa dạng và xuất hiện trong nhiều lĩnh vực, bao gồm:

1. Nhiếp ảnh:RGB được sử dụng trong các hệ thống nhiếp ảnh màu, như việc kết hợp ba ảnh có bộ lọc màu (đỏ, xanh lá và xanh dương) để tạo ra một bức ảnh màu hoàn chỉnh. Các phương pháp này được áp dụng trong các quá trình như Autochrome Lumière, Joly color screen, và các công nghệ màn hình khác từ đầu thế kỷ 20.
Hệ màu RGB cũng được sử dụng trong máy ảnh kỹ thuật số hiện đại, với các cảm biến ảnh sử dụng bộ lọc Bayer để tái tạo hình ảnh màu.

2. RGB là nền tảng của các hệ thống truyền hình màu, bắt đầu từ các thử nghiệm cơ học của John Logie Baird vào những năm 1920. Các hệ thống màu sau này (như của CBS) sử dụng bánh xe màu RGB quay để tạo ra màu sắc.
Công nghệ RGB được áp dụng trong các màn hình TV và máy chiếu hiện đại, đặc biệt là trên các màn hình CRT, LCD, OLED và Plasma.


3. Trong các máy tính cá nhân, RGB được sử dụng trong các card đồ họa để hiển thị màu sắc trên màn hình. Các chuẩn màn hình như VGA (Video Graphics Array) và Super VGA sử dụng tín hiệu analog RGB để cung cấp màu sắc chân thực.
RGB24 và RGB32 là các chuẩn màu thông dụng trong các card đồ họa và file hình ảnh, với RGB24 có thể hiển thị tới 16,777,216 màu khác nhau.

## 5. Provide the mathematical definition of a tensor.

Khi dữ liệu biểu diễn dạng 1 chiều, người ta gọi là vector, mặc định khi viết vector sẽ viết dưới dạng cột.

Khi dữ liệu dạng 2 chiều, người ta gọi là ma trận, kích thước là số hàng * số cột. Ví dụ:

*Vector v kích thước n, ma trận W kích thước m x n*

\begin{align}
v &=
\begin{bmatrix}
v_1 \\
v_2 \\
\vdots \\
v_n
\end{bmatrix}
&
W &=
\begin{bmatrix}
 w_{11} & x_{12}& ... & x_{1n}  \\
w_{21} & x_{22}& ...  & x_{2n} \\
... & ... & ... \\
w_{m1} & x_{m2}& ...  & x_{mn}
\end{bmatrix}
\end{align}


Khi dữ liệu có nhiều hơn 2 chiều, nó sẽ được gọi là tensor. Ví dụ, dữ liệu 3 chiều sẽ được gọi là tensor 3D, dữ liệu 4 chiều là tensor 4D, v.v. Có thể nhận thấy rằng ma trận thực chất là sự kết hợp của các vector có cùng kích thước. Khi xếp n vector có kích thước m cạnh nhau, ta sẽ có một ma trận có kích thước m × n. Tương tự, tensor 3 chiều là sự kết hợp của các ma trận có cùng kích thước, và khi xếp k ma trận có kích thước m × n chồng lên nhau, ta sẽ có một tensor có kích thước m × n × k.Như vậy, tensor tổng quan trong toán học cũng chỉ là một cách để biểu diễn sự kết hợp khi xếp các ma trận chồng lên nhau.

Công thức toán học tổng quát của tensor n chiều có thể được diễn tả như sau:

Một tensor n chiều có thể được mô tả bằng một mảng có $n$ chỉ số, và mỗi chỉ số chỉ ra một chiều của tensor. Một tensor với kích thước $(d_1, d_2, ..., d_n)$ có $n$ chỉ số, mỗi chỉ số có giá trị từ 1 đến $d_i$ (với $i = 1, 2, ..., n$).

Công thức tổng quát của một tensor có $n$ chiều sẽ là:

$$T = T(i_1, i_2, ..., i_n)$$

Trong đó:

* $T$ là tensor có $n$ chiều.
* $i_1, i_2, ..., i_n$ là các chỉ số (index) của tensor, với $i_k \in \{1, 2, ..., d_k\}$ cho mỗi chiều $k$ từ 1 đến $n$.
* $d_k$ là kích thước của chiều thứ $k$.


Ví dụ:

* **Tensor 1D (vector)** có kích thước $(d_1)$, với chỉ số $T(i_1)$, trong đó $i_1 \in \{1, 2, ..., d_1\}$.
* **Tensor 2D (ma trận)** có kích thước $(d_1, d_2)$, với chỉ số $T(i_1, i_2)$, trong đó $i_1 \in \{1, 2, ..., d_1\}$ và $i_2 \in \{1, 2, ..., d_2\}$.
* **Tensor 3D** có kích thước $(d_1, d_2, d_3)$, với chỉ số $T(i_1, i_2, i_3)$, trong đó $i_1 \in \{1, 2, ..., d_1\}$, $i_2 \in \{1, 2, ..., d_2\}$, và $i_3 \in \{1, 2, ..., d_3\}$.

Một tensor 3D có thể được coi là một danh sách ma trận ba chiều:

$$
\begin{bmatrix}
\begin{bmatrix}
x_{0,0} & x_{0,1} & x_{0,2} \\
x_{1,0} & x_{1,1} & x_{1,2} \\
x_{2,0} & x_{2,1} & x_{2,2}
\end{bmatrix} &
\begin{bmatrix}
x_{0,0} & x_{0,1} & x_{0,2} \\
x_{1,0} & x_{1,1} & x_{1,2} \\
x_{2,0} & x_{2,1} & x_{2,2}
\end{bmatrix} &
\begin{bmatrix}
x_{0,0} & x_{0,1} & x_{0,2} \\
x_{1,0} & x_{1,1} & x_{1,2} \\
x_{2,0} & x_{2,1} & x_{2,2}
\end{bmatrix}
\end{bmatrix}
=
\begin{bmatrix}
\begin{bmatrix}
\vec{x}_0 \\
\vec{x}_1 \\
\vec{x}_2
\end{bmatrix}&
\begin{bmatrix}
\vec{x}_0 \\
\vec{x}_1 \\
\vec{x}_2
\end{bmatrix}&
\begin{bmatrix}
\vec{x}_0 \\
\vec{x}_1 \\
\vec{x}_2
\end{bmatrix}
\end{bmatrix}
=
\begin{bmatrix}
X_0 &X_1&X_2 \\
\end{bmatrix}
= \mathcal{X}
$$

Các phép toán cơ bản của Tensor: vì bản chất Tensor là sự kết hợp của ma trận nên tensor thừa kế những phép toán cơ bản của ma trận như: cộng, trừ, nhân 1 số nguyên với ma trận, nhân 2 ma trận, chia, nghịch đảo, chuyển vị.

Các phép toán trên tensor 3 chiều (hoặc tensor bậc 3) tương tự như các phép toán trên tensor bậc cao nói chung, nhưng với các đặc thù của một tensor 3D. Dưới đây là các phép toán cơ bản và công thức tổng quát:

1. Cộng tensor (Tensor Addition):
- Điều kiện: Các tensor phải có cùng kích thước.
- Công thức tổng quát: $C_{ijk} = A_{ijk} + B_{ijk}$
trong đó $A$, $B$, và $C$ là các tensor 3 chiều với chỉ số $i$, $j$, và $k$ chạy qua các chiều của tensor.
2. Nhân tensor với một hằng số (Scalar Multiplication):
- Công thức: $C_{ijk} = \alpha \cdot A_{ijk}$, trong đó $\alpha$là hằng số, và $A$, $C$ là các tensor 3 chiều.
3. Nhân hai tensor 3 chiều (Tensor Multiplication):
- Đây là phép toán phức tạp hơn, và có thể có nhiều dạng khác nhau như nhân điểm (element-wise) hoặc nhân ma trận.
- Nhân điểm (Element-wise multiplication):
  - Công thức: $C_{ijk} = A_{ijk} \cdot B_{ijk}$, các phần tử của tensor $A$ và $B$ được nhân với nhau một cách trực tiếp.
- Nhân ma trận (Matrix multiplication): Giả sử mỗi "lát cắt" của tensor 3D có thể coi là một ma trận 2D. Phép nhân ma trận có thể được áp dụng giữa các lát cắt (slices) này. Phép nhân ma trận giữa $A$ và $B$ sẽ tạo ra tensor $C$ có kích thước $(I, J, L)$. Mỗi phần tử của tensor $C$ được tính như sau:

$$C_{ijl} = \sum_{k=1}^{K} A_{ijk} \cdot B_{kl}$$
4. Chuyển vị (Transpose):
- Tương tự như chuyển vị ma trận, chuyển vị của tensor 3 chiều có thể thực hiện bằng cách hoán đổi các chỉ số.
- Ví dụ, nếu $A_{ijk}$ là một tensor, thì chuyển vị có thể là $A_{kji}$ (hoán đổi các chỉ số $i$, $j$, $k$).

5. Hợp nhất các chiều (Tensor Flattening):
- Chuyển đổi tensor 3D thành tensor 1D hoặc 2D, tức là "phẳng hóa" tensor.
- Ví dụ, nếu bạn có một tensor có kích thước $(I, J, K)$ bạn có thể chuyển nó thành một vector có kích thước $(I x J x K).

6. Phép toán theo trục (Slicing and Indexing):
- Phép toán này giúp trích xuất các lát cắt (slice) của tensor theo các chỉ số.
- Ví dụ, trích xuất lát cắt trên trục $i$ từ tensor $A$ có thể viết là $A[i, :, :]$ tức là lấy phần tử thứ $i$ trên trục đầu tiên và giữ nguyên các phần tử ở các trục còn lại.

```python
def get_pixel_intensity(image_path, i, j):
    """
    Reads a grayscale image using OpenCV and retrieves the intensity
    at the specified pixel coordinates (i, j).

    Args:
        image_path (str): Path to the grayscale image.
        i (int): Row index of the pixel (y-coordinate).
        j (int): Column index of the pixel (x-coordinate).

    Returns:
        int: Pixel intensity (0-255) if the coordinates are valid.
        str: Error message if the coordinates are out of bounds.
    """
    # Read the image in grayscale mode
    image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

    if image is None:
        return "Image could not be loaded. Please check the file path."

    # Check if (i, j) is within the image bounds
    if 0 <= i < image.shape[0] and 0 <= j < image.shape[1]:
        return image[i, j]  # Return the pixel intensity
    else:
        return "Coordinates are out of bounds."

```

```python
!pip install opencv-python

import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


def inpaint_image(image_path, mask_path, inpaint_radius=3, inpaint_method=cv.INPAINT_TELEA):
    """
    Applies inpainting to an image using a specified mask.

    Parameters:
        image_path (str): Path to the input image.
        mask_path (str): Path to the grayscale mask image.
        inpaint_radius (int): Radius of a circular neighborhood of each point inpainted.
        inpaint_method (int): Inpainting method, either cv.INPAINT_TELEA or cv.INPAINT_NS.

    Returns:
        dst (ndarray): The inpainted image.
    """
    img = cv.imread(image_path)
    # Load the mask image
    mask = cv.imread(mask_path, cv.IMREAD_GRAYSCALE)  # Load mask as grayscale

    if img is None:
        raise ValueError(f"Cannot load image from path: {image_path}")

    # Resize the mask to match the image dimensions
    mask = cv.resize(mask, (img.shape[1], img.shape[0])) # Resizes mask to match img

    # Use the loaded mask (NumPy array) in cv.inpaint
    dst = cv.inpaint(img, mask, inpaint_radius, inpaint_method)
    return dst
```

```python
def threshold_image(image, I_lower, I_upper, region):
    """
    Apply a threshold to a specific region of an image.
    Pixels within the range [I_lower, I_upper] will be set to 255, otherwise set to 0.
    Pixels outside the specified region will remain unchanged.

    Args:
        image (numpy.ndarray): Input image as a 2D or 3D array.
        I_lower (float): Lower threshold value.
        I_upper (float): Upper threshold value.
        region (tuple, optional): Coordinates of the region (x_start, y_start, x_end, y_end).
                                  If None, apply to the entire image.

    Returns:
        numpy.ndarray: Image after applying the threshold.
    """
    # Copy the image to avoid modifying the original
    result_image = np.copy(image)

    if region:
        x_start, y_start, x_end, y_end = region
        # Apply the threshold only within the specified region
        result_image[y_start:y_end, x_start:x_end] = np.where(
            (I_lower <= image[y_start:y_end, x_start:x_end]) &
            (image[y_start:y_end, x_start:x_end] <= I_upper),
            255,
            0
        )
        result_image[np.r_[:y_start, y_end:result_image.shape[0]], :] = 0
        result_image[:, np.r_[:x_start, x_end:result_image.shape[1]]] = 0
    else:
        # Apply the threshold to the entire image
        result_image = np.where((I_lower <= image) & (image <= I_upper), 255, 0)

    return result_image
```

```python
def mark_multiple_points_on_image(image_path, points):
    """
    Marks multiple points with different colors on the grayscale image and displays a legend.

    Args:
        image_path (str): Path to the image file.
        points (list): List of tuples, where each tuple represents a point (i, j).

    Returns:
        None: Displays the modified image with the marked points and legend.
    """
    # Load the grayscale image
    image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

    if image is None:
        print("Error: Unable to load the image. Please check the file path.")
        return

    # Define colors and labels for the points
    colors = ['red', 'green', 'blue', 'yellow']
    labels = [f"Point {i+1}: ({points[i][0]},{points[i][1]})" for i in range(len(points))]

    # Check if all points are within the bounds of the image
    for i, j in points:
        if not (0 <= i < image.shape[0] and 0 <= j < image.shape[1]):
            print("Error: Some coordinates are out of bounds.")
            return

    # Mark the points on the image with different colors
    plt.imshow(image, cmap='gray')
    for i, (x, y) in enumerate(points):
        plt.scatter([y], [x], c=colors[i % len(colors)], s=10, label=labels[i])

    # Display the legend on the left
    plt.legend(loc="best", bbox_to_anchor=(0, 0.5), fontsize=7)

    plt.title("Image with Marked Points")
    plt.show()
```

Gọi $I(a, j)$ là điểm nằm trong vùng object. Thì ta có, $$I(a, j) \in (I_0 - \epsilon, I_0 + \epsilon)$$

$\forall a \in [i_0, i_1]$

Với object đầu tiên, ta có $i_0 = 64$ và $i_1 = 82$.

Thì $I_1 = \{I(140,64), I(140,65), ... I(140,82)\}$ (ở đây ta chọn bước nhảy = 1 để độ mịn được mịn nhất có thể)

```python
def find_max_intensity(image_path, start_row, end_row, column):
    """
    Tìm giá trị cường độ pixel lớn nhất trong một phạm vi hàng đã cho.

    Args:
    image_path: Đường dẫn đến hình ảnh.
    start_row: Hàng bắt đầu.
    end_row: Hàng kết thúc.
    column: Cột cần kiểm tra.

    Returns:
    Giá trị cường độ lớn nhất và một danh sách các cường độ.
    """
    intensities = []
    for row in range(start_row, end_row + 1):
        intensity = get_pixel_intensity(image_path, column, row)
        intensities.append(intensity)

    max_intensity = max(intensities)
    min_intensity = min(intensities)

    return max_intensity, intensities, min_intensity



max_intensity, intensities, min_intensity = find_max_intensity(caucasian_path, 64, 82, 140)

print("Danh sách cường độ:", intensities)
print("Giá trị cường độ lớn nhất:", max_intensity)
print("Giá trị cường độ bé nhất:", min_intensity)
```

Vì vậy, $sup(I_1) = 236$ và $inf(I_1) = 222$. Vậy $I_0 = 229$ và $\epsilon = 7$. Vậy $ I_1 \in [222, 236]$
