## 1. What are the main ideas in deep learning for image processing?


Deep Learning, hay còn gọi là học sâu, là một nhánh của trí tuệ nhân tạo sử dụng mạng nơ-ron thần kinh (neural network) để tìm hiểu cách biểu diễn hữu ích của các tính năng trực tiếp từ dữ liệu. Mạng nơ-ron này được lấy cảm hứng từ cấu trúc và chức năng của não bộ con người, cho phép máy tính "học" từ dữ liệu đầu vào và tự động trích xuất các đặc trưng quan trọng. Một trong những ứng dụng nổi bật của Deep Learning là trong lĩnh vực xử lý ảnh. Deep Learning có khả năng tự động học cách biểu diễn các đặc trưng hình ảnh phức tạp, từ đó giúp giải quyết nhiều bài toán xử lý ảnh hiệu quả.

Một ví dụ điển hình là việc sử dụng mạng nơ-ron được huấn luyện trước để xác định và loại bỏ các thành phần nhiễu (noise) khỏi hình ảnh. Mạng nơ-ron này được huấn luyện trên một tập dữ liệu lớn gồm các hình ảnh nhiễu và hình ảnh sạch tương ứng. Sau khi huấn luyện, mạng nơ-ron có thể tự động phát hiện và loại bỏ nhiễu trong các hình ảnh mới, giúp cải thiện chất lượng hình ảnh đáng kể.

Deep Learning đã làm thay đổi hoàn toàn việc xử lý hình ảnh, tạo điều kiện cho những tiến bộ từng được cho là khoa học viễn tưởng. Từ chẩn đoán bệnh đến cung cấp năng lượng cho ô tô tự lái, nâng cao hiệu ứng hình ảnh và cải thiện hệ thống an ninh, các ứng dụng của deep learning rất rộng lớn và đa dạng. Công nghệ tiếp tục phát triển, hứa hẹn nhiều giải pháp và ứng dụng sáng tạo hơn nữa trong tương lai.

## 3. Why are Convolutional Neural Networks (CNNs) commonly used in image processing?

Mạng Nơ-ron Tích chập (CNN) được sử dụng rộng rãi trong phân loại ảnh nhờ vào những ưu điểm chính khiến chúng đặc biệt hiệu quả so với các thuật toán khác:

Học Đặc Trưng Phân Cấp
Kết Nối Cục Bộ: CNN sử dụng các lớp tích chập áp dụng các bộ lọc lên các vùng cục bộ của hình ảnh đầu vào. Điều này giúp chúng nắm bắt được các cấp độ không gian của đặc trưng, từ các cạnh và họa tiết đơn giản trong các lớp đầu cho tới các hình dạng và đối tượng phức tạp hơn trong các lớp sâu hơn. Tự Động Trích Xuất Đặc Trưng: Không giống như các thuật toán truyền thống yêu cầu trích xuất đặc trưng thủ công, CNN tự động học các đặc trưng tốt nhất cho việc phân loại thông qua quá trình huấn luyện.

Chia Sẻ Tham Số
Giảm Độ Phức Tạp: CNN sử dụng cùng một bộ lọc (kernel) trên các phần khác nhau của hình ảnh, điều này làm giảm đáng kể số lượng tham số so với các mạng kết nối đầy đủ. Điều này giúp chúng ít bị quá khớp hơn và có khả năng tổng quát hóa tốt hơn trên dữ liệu chưa thấy.

Tính Bất Biến Dịch Chuyển
Khả Năng Chống Lệch: CNN thể hiện khả năng chống dịch chuyển, nghĩa là chúng có thể nhận diện đối tượng trong ảnh bất kể vị trí của chúng. Điều này đạt được thông qua các lớp pooling giảm kích thước không gian, giữ lại các đặc trưng quan trọng trong khi loại bỏ thông tin kém quan trọng.

Tính Toán Hiệu Quả
Tích Chập vs. Lớp Kết Nối Đầy Đủ: Chi phí tính toán của các phép tích chập thấp hơn lớp kết nối đầy đủ, đặc biệt đối với dữ liệu có độ phân giải cao như hình ảnh. Sự hiệu quả này cho phép CNN mở rộng tốt hơn với các tập dữ liệu lớn.

Thành Công Trong Thực Tế
Hiệu Suất Thực Nghiệm: CNN liên tục vượt trội hơn các phương pháp truyền thống (như SVMs, cây quyết định hay phương pháp đặc trưng thủ công) trên nhiều tiêu chuẩn phân loại hình ảnh (ví dụ: ImageNet) nhờ vào khả năng học các mẫu phức tạp.

Học Chuyển Tiếp (Transfer Learning)
Sử Dụng Mô Hình Đã Huấn Luyện Trước: CNN có thể tận dụng học chuyển tiếp, trong đó các mô hình đã được huấn luyện trên các tập dữ liệu lớn có thể được tinh chỉnh cho các tác vụ cụ thể với tập dữ liệu nhỏ hơn. Điều này đặc biệt hữu ích trong các tình huống khan hiếm dữ liệu đã gán nhãn.

Học Từ Đầu Đến Cuối
Quy Trình Đơn Giản: CNN cho phép một quá trình học từ đầu đến cuối, nơi dữ liệu pixel thô được đưa trực tiếp vào mạng và mô hình học để tối ưu hóa cho việc phân loại mà không cần các bước trung gian.

## 4. What is the difference between Convolutional Neural Networks (CNNs) and Fully Connected Neural Networks in image processing?

Mạng kết nối hoàn toàn, thường được gọi là mạng nơ-ron dày đặc, là dạng truyền thống của mạng nơ-ron nơi mỗi nơ-ron trong một lớp được kết nối với mọi nơ-ron trong lớp tiếp theo. Các mạng này thường chỉ bao gồm các trọng số ở dạng số thực kết nối các nơ-ron và hoạt động xuất sắc cho các tác vụ mà các đầu vào là độc lập và không có cấu trúc vốn có, chẳng hạn như dữ liệu bảng.

Ngược lại, Mạng Nơ-ron Tích chập (CNN) được thiết kế đặc biệt để xử lý các đầu vào có cấu trúc dạng lưới, điều này đặc biệt đúng với hình ảnh. Hình ảnh vốn dĩ là các cấu trúc lưới hai chiều mà các pixel có liên hệ về không gian với nhau. CNN tận dụng cấu trúc không gian này bằng cách chứa ít nhất một lớp tích chập thực hiện các phép tích chập. Các phép toán này sử dụng các bộ lọc kernel làm trọng số, có cấu trúc phức tạp hơn so với các trọng số dạng số thực đơn lẻ được sử dụng trong các mạng kết nối hoàn toàn.

Việc sử dụng các kernel làm trọng số thay vì các trọng số đơn giúp CNN học được các cấp độ không gian của đặc trưng, như việc phát hiện các cạnh, họa tiết và các hình dạng phức tạp. Các lớp tích chập trượt các bộ lọc này qua dữ liệu đầu vào, thực hiện các tích chập giữa các phần của bộ lọc và đầu vào, từ đó nắm bắt các mẫu không gian cục bộ. Phương pháp này giúp CNN tự động học các cấp độ không gian của đặc trưng, từ các cạnh và họa tiết đơn giản trong các lớp tích chập ban đầu đến các mẫu và phần đối tượng phức tạp ở các lớp sâu hơn.

Hơn nữa, CNN đảm bảo mối quan hệ không gian trong dữ liệu do chính phép toán tích chập và khả năng chia sẻ tham số vốn có của các mạng này. Điều này bao gồm việc sử dụng cùng một kernel trên toàn bộ hình ảnh, cho phép nhận diện đặc trưng bất kể vị trí của chúng trong hình ảnh.

Tuy nhiên, các mạng kết nối hoàn toàn yêu cầu mỗi đầu vào, chẳng hạn như một hình ảnh, được làm phẳng thành một vector một chiều trước khi xử lý. Quá trình làm phẳng này làm mất đi thông tin về không gian và mối liên hệ cục bộ của dữ liệu đầu vào, do đó bỏ qua các mối quan hệ giữa các pixel cung cấp ngữ cảnh quan trọng cần thiết cho các nhiệm vụ nhận diện.

Việc bảo toàn không gian này rất quan trọng trong phân loại hình ảnh và các nhiệm vụ liên quan đến thị giác khác, nơi việc phân biệt các mẫu và cấu trúc ở các tỷ lệ và vị trí khác nhau là quan trọng. Nó cho phép CNN nắm bắt và học từ các phức tạp hiện diện trong dữ liệu hình ảnh hiệu quả hơn nhiều, thường dẫn đến hiệu suất vượt trội so với các mạng kết nối hoàn toàn, vốn không tận dụng được các động lực không gian quan trọng này.

## 5. What is a Kernel (or Filter) in CNNs? How does it work?

Một Filter, hay còn gọi là kernel, trong CNNs là một ma trận nhỏ của các trọng số trượt qua dữ liệu đầu vào (chẳng hạn như một hình ảnh), thực hiện phép nhân từng phần tử với phần của dữ liệu đầu vào mà nó đang bao phủ, sau đó cộng tất cả các kết quả để tạo ra một pixel đầu ra duy nhất. Quá trình này được gọi là tích chập.

Bộ lọc là thành phần cốt lõi làm cho CNN hoạt động hiệu quả. Chúng là thành phần chính giúp mô hình trích xuất các đặc trưng hữu ích từ dữ liệu đầu vào.

Dưới đây là cách nó hoạt động:

Kích thước: Bộ lọc thường là các ma trận nhỏ, hình vuông. Kích thước của bộ lọc là một siêu tham số quan trọng mà bạn cần thiết lập. Những kích thước thông thường bao gồm 3x3, 5x5 và 7x7.

Sliding Bộ lọc: Bộ lọc trượt qua dữ liệu đầu vào, di chuyển một số pixel nhất định mỗi lần, được xác định bởi "bước nhảy" (stride) (mà chúng ta sẽ thảo luận sau). Tại mỗi vị trí, bộ lọc bao phủ một phần nhỏ của hình ảnh đầu vào.

Phép toán Tích chập: Tại mỗi vị trí, bộ lọc thực hiện một phép toán trên phần của hình ảnh mà nó đang bao phủ. Phép toán này bao gồm nhân từng phần tử của trọng số bộ lọc với các giá trị pixel trong hình ảnh, sau đó cộng các kết quả này lại. Điều này tạo ra một giá trị đơn lẻ đại diện cho vị trí cụ thể đó trên bản đồ đặc trưng đầu ra.

Learning: Trong quá trình huấn luyện, CNN học các giá trị tốt nhất cho các trọng số của bộ lọc để đạt được nhiệm vụ cụ thể (ví dụ: phân loại hình ảnh, phát hiện đối tượng, v.v.). Điều này được thực hiện thông qua lan truyền ngược và các thuật toán tối ưu hóa, giống như bất kỳ trọng số nào khác trong một mạng nơ-ron.

Trích xuất Đặc trưng: Bộ lọc chịu trách nhiệm trích xuất đặc trưng trong CNN. Ví dụ, một số bộ lọc có thể trở nên chuyên biệt cho việc phát hiện các cạnh ngang trong một hình ảnh, các bộ lọc khác có thể phát hiện cạnh dọc, màu sắc, họa tiết, v.v. Khi mô hình trở nên sâu hơn, các bộ lọc có thể nhận biết các mẫu phức tạp hơn.

**Bằng cách áp dụng nhiều bộ lọc cho dữ liệu đầu vào, một CNN có thể học cách phát hiện một loạt các đặc trưng phong phú. Điều này làm cho chúng trở thành công cụ mạnh mẽ cho các công việc xử lý hình ảnh.**

## 6. Why are pooling operations (e.g., MaxPooling, AveragePooling) important in image processing models?

Pooling quan trọng trong Image Processing vì nó giảm phương sai, giảm độ phức tạp tính toán (vì pooling tối đa/trung bình 2x2 có thể giảm 75% dữ liệu) và trích xuất các đặc trưng cấp thấp từ vùng lân cận, góp phần tăng cường tính ổn định và độ bền của mô hình. Ngoài ra, Cả MaxPooling và AveragePooling đều giúp giảm kích thước các bản đồ đặc trưng. Điều này dẫn đến việc giảm số lượng tham số và tính toán cần thiết trong mạng, làm cho mô hình hiệu quả hơn.

Pooling cho phép mạng tập trung vào các đặc trưng quan trọng bằng cách giảm bớt các chi tiết kém quan trọng hơn. Pooling tối đa có thể trích xuất các đặc trưng nổi bật như cạnh, trong khi pooling trung bình cung cấp một cái nhìn tổng quát hơn về các đặc trưng. Bằng cách giảm kích thước một cách khôn ngoan, pooling làm mô hình ít nhạy hơn với các nhiễu loạn nhỏ hay sự thay đổi trong dữ liệu đầu vào.


## 7. Explain the roles of Conv2D, BatchNormalization, and Activation layers in image processing networks.

**Conv2D**

Phép tích chập liên quan đến các tín hiệu một chiều được gọi là tích chập 1D hoặc chỉ đơn giản là tích chập. Ngược lại, nếu phép tích chập được thực hiện giữa hai tín hiệu kéo dài theo hai chiều vuông góc với nhau (tức là nếu tín hiệu có bản chất hai chiều), thì nó sẽ được gọi là tích chập 2D. Khái niệm này có thể được mở rộng để bao gồm các tín hiệu đa chiều, do đó chúng ta có thể có tích chập đa chiều.

Trong lĩnh vực số hóa, tích chập được thực hiện bằng cách nhân và tích lũy các giá trị tức thời của các mẫu chồng lấn tương ứng với hai tín hiệu đầu vào, trong đó một tín hiệu được lật ngược. Định nghĩa này của tích chập 1D cũng áp dụng cho tích chập 2D, ngoại trừ trong trường hợp sau, một trong các đầu vào được lật ngược hai lần.

Loại phép toán này được sử dụng rộng rãi trong lĩnh vực xử lý ảnh số, trong đó ma trận 2D đại diện cho hình ảnh sẽ được tích chập với một ma trận nhỏ hơn gọi là kernel 2D.

**BatchNormalization**

Batch Norm là một kỹ thuật bình thường hóa thực hiện giữa các lớp của Mạng Nơ-ron, không phải trên dữ liệu thô. Nó hoạt động trên các mini-batches thay vì toàn bộ tập dữ liệu, giúp tăng tốc độ huấn luyện và cho phép sử dụng các tỷ lệ học cao hơn.

Batch Norm chuẩn hóa đầu ra của các neuron, giúp ổn định phân phối đầu vào của các lớp khác trong mạng. Điều này làm giảm sự thay đổi trong phân phối dữ liệu đầu vào của các lớp bên trong, giúp quá trình học tập diễn ra thuận lợi hơn

Trong CNN, Batch Norm hoạt động theo cách tương tự nhưng phải tuân theo tính chất tích chập. Các bộ lọc được chia sẻ mà đi qua các bản đồ đặc trưng (feature maps) của đầu vào. Batch Norm chuẩn hóa các bản đồ đầu ra theo cùng cách, chia sẻ trên các bản đồ đặc trưng.

Các tham số được sử dụng để chuẩn hóa được tính cho từng bản đồ đặc trưng toàn bộ, với mỗi bản đồ đặc trưng có một trung bình và độ lệch chuẩn duy nhất được sử dụng trên tất cả các đặc trưng mà nó chứa.

**Activation Layer**
Trong các mạng xử lý hình ảnh, các lớp kích hoạt đóng vai trò quan trọng trong việc học và trích xuất các mối quan hệ phi tuyến tính từ dữ liệu. Việc đưa phi tuyến tính vào mô hình là cần thiết vì dữ liệu trong thế giới thực thường không tuyến tính. Nếu không có các lớp kích hoạt phi tuyến tính, mạng nơ-ron sâu với nhiều lớp sẽ chỉ hoạt động như một mô hình tuyến tính, hạn chế khả năng mô phỏng các mối quan hệ phức tạp trong dữ liệu. Do đó, các lớp kích hoạt giúp mạng lưới học các biểu diễn phi tuyến tính quan trọng cho việc nhận diện và phân loại hình ảnh.

Các hàm kích hoạt phổ biến trong các mạng xử lý hình ảnh bao gồm ReLU (Rectified Linear Unit), Sigmoid, và Tanh. ReLU là một trong những hàm được ưa chuộng nhất vì sự đơn giản và hiệu quả của nó trong việc giảm vấn đề biến mất đạo hàm. Nó giữ cho đầu ra của các lớp trọng số không bị âm và nhanh chóng tính toán. Ngoài ra, các biến thể của ReLU như Leaky ReLU và Parametric ReLU được phát triển để giải quyết vấn đề các đơn vị bị chết của ReLU tiêu chuẩn bằng cách cho phép một gradient khác không khi tín hiệu vào là âm.

Các lớp kích hoạt cũng đóng vai trò quan trọng trong việc trích xuất các đặc trưng từ dữ liệu. Chúng cho phép mạng học được các biểu diễn phức tạp của dữ liệu đầu vào, giúp quá trình học tập nhanh hơn và hiệu quả hơn. Ví dụ, ReLU giúp giảm thiểu bão hòa bằng cách không giới hạn đầu ra vào một phạm vi cụ thể, điều này duy trì các gradient mạnh mẽ cho các lớp ở trên. Sự lựa chọn đúng hàm kích hoạt dựa trên dữ liệu và mục tiêu của mô hình có thể tối ưu hóa hiệu suất và tính chính xác của các mạng xử lý hình ảnh.

## 8. What problem do skip connections in architectures like U-Net or ResNet help solve? Why are they important?

Bỏ qua kết nối (skip connection) trong các kiến trúc mạng sâu như U-Net và ResNet mang lại nhiều lợi ích quan trọng, đặc biệt là trong việc giải quyết vấn đề biến mất đạo hàm (vanishing gradient problem). Trong các mạng nơ-ron sâu, khi tín hiệu sai số được truyền ngược lại qua các lớp, giá trị gradient có xu hướng trở nên rất nhỏ, dẫn đến khó khăn trong việc huấn luyện các lớp phía dưới. Bỏ qua kết nối cho phép tín hiệu sai số truyền trực tiếp qua nhiều lớp mà không bị làm suy yếu, đảm bảo rằng gradient vẫn đủ lớn để quá trình tối ưu hóa tiếp tục hiệu quả. Điều này đặc biệt quan trọng trong các mạng rất sâu như ResNet, nơi mà độ sâu của mô hình có thể lên tới hàng trăm lớp.

Ngoài việc giải quyết vấn đề biến mất đạo hàm, bỏ qua kết nối cũng hỗ trợ học tập các biểu diễn phức tạp hơn bằng cách cho phép truy cập trực tiếp từ các tầng phía sau đến các đặc trưng từ các tầng phía trước. Ví dụ, trong kiến trúc U-Net, được sử dụng rộng rãi trong tác vụ phân đoạn ảnh y tế, các thông tin chi tiết của hình ảnh từ các tầng đi xuống được kết nối trực tiếp với các tầng đi lên tương ứng. Điều này cho phép mô hình tái sử dụng thông tin không gian quan trọng, giúp cải thiện độ chính xác của kết quả phân đoạn.

Việc sử dụng bỏ qua kết nối cũng thúc đẩy khả năng tổng quát hóa của mô hình (khả năng lắp ghép). Trong ResNet, các kết nối thẳng cho phép các lớp học một hàm đồng nhất nếu cần thiết; có nghĩa là chúng có thể học để chuyển tuyến tính từ đầu vào đến đầu ra mà không cần tạo ra các biến đổi phức tạp. Điều này đặc biệt hữu ích trong việc giảm thiểu overfitting, giúp mô hình dự đoán chính xác hơn trên dữ liệu chưa từng thấy. Ví dụ, ResNet đã đạt được thành công lớn trong cuộc thi ImageNet, thể hiện khả năng tổng quát hóa vượt trội so với các mạng không sử dụng bỏ qua kết nối.

Một ví dụ cụ thể về lợi ích của bỏ qua kết nối là trong lĩnh vực xử lý ảnh y tế, nơi U-Net đã cho thấy khả năng phân đoạn các cấu trúc giải phẫu với độ chính xác cao. Các kết nối bỏ qua trong U-Net giúp giữ lại các thông tin quan trọng từ các lớp ban đầu và kết hợp với các đặc trưng trừu tượng từ các lớp sâu hơn, tạo ra một kết quả phân đoạn chính xác và chi tiết hơn. Trong ResNet, bỏ qua kết nối đã cho phép xây dựng các mạng rất sâu mà vẫn có thể huấn luyện hiệu quả, điều này đã cách mạng hóa khả năng xử lý và nhận diện ảnh trong nhiều ứng dụng thực tiễn.



## 9. Compare popular CNN architectures like AlexNet, VGG, and ResNet in image processing.

Trong lĩnh vực xử lý ảnh, các kiến trúc CNN nổi tiếng như AlexNet, VGG, và ResNet đều đã đóng góp quan trọng trong việc cải thiện hiệu suất và độ chính xác của mô hình. AlexNet là một trong những mạng đầu tiên đạt được độ chính xác đáng kể trên thử thách ImageNet năm 2012 nhờ sử dụng các lớp tích chập sâu và phương pháp tăng cường dữ liệu để giảm overfitting. AlexNet sử dụng Rectified Linear Unit (ReLU) để giải quyết vấn đề biến mất đạo hàm và áp dụng các lớp Local Response Normalization (LRN) cùng với dropout để giảm overfitting, giúp mạng học nhanh hơn và ổn định hơn.

VGGNet, ra đời sau AlexNet, tập trung vào việc giảm số lượng tham số trong các lớp tích chập bằng cách sử dụng các kernel kích thước cố định 3x3. Các biến thể như VGG16 và VGG19 nổi bật với số lượng lớp lớn hơn, cho phép mô hình trích xuất các đặc trưng chi tiết và chính xác hơn mà không làm tăng đáng kể số lượng tham số. Việc kết hợp nhiều lớp tích chập 3x3 đã chứng minh rằng số lượng biến huấn luyện có thể được giảm mà vẫn duy trì được khả năng mô phỏng các đặc trưng không gian sâu.


ResNet đã cách mạng hóa CNN với việc giới thiệu các skip connection, giải quyết hiệu quả vấn đề biến mất đạo hàm và cho phép mạng trở nên rất sâu. Những kết nối này cho phép mạng học một ánh xạ đồng nhất dễ dàng hơn, đảm bảo rằng độ sâu tăng thêm không làm giảm hiệu suất mô hình. ResNet sử dụng block dư (Residual blocks) để hỗ trợ việc học các ánh xạ phức tạp hơn mà không làm tăng độ phức tạp tính toán hoặc gặp các vấn đề về gradient như trong các mạng truyền thống. Các kiến trúc như ResNet50 và ResNet101 thể hiện khả năng tổng quát hóa tuyệt vời trên dữ liệu chưa thấy và là những tiêu chuẩn phổ biến trong nghiên cứu và ứng dụng thực tiễn.

```python
def add_impulse_noise(image, prob):
    """
    Adds impulse (salt & pepper) noise to an image.

    Args:
    - image: Input image (numpy array).
    - prob: Probability of adding noise (0 < prob < 1).

    Returns:
    - noisy_img: Image with added noise.
    """
    # Create a copy of the original image to avoid modifying it directly
    noisy_img = image.copy()
    # Generate a random matrix with the same dimensions as the image
    random_matrix = np.random.rand(*image.shape[:2])

    # Add "salt" noise (white pixels)
    noisy_img[random_matrix < prob/2] = 255

    # Add "pepper" noise (black pixels)
    noisy_img[random_matrix > 1 - prob/2] = 0

    return noisy_img

```

```python
import os
import numpy as np
from skimage.transform import rescale, resize
import tensorflow as tf
from tqdm.keras import TqdmCallback
from skimage import data  # Importing sample data module from skimage
from matplotlib import pyplot as plt
```

```python
def dip_workflow(x0,
                 x_true,
                 f,
                 f_input_shape,
                 z_std=0.1,
                 loss_mask=None,
                 num_iters=1,
                 init_lr=0.01,
                 save_filepath=None):
    """Deep Image Prior (DIP) workflow for image restoration tasks.
    Args:
        x0: Input image (e.g., noisy or corrupted image).
        x_true: Ground-truth image used for metrics evaluation.
        f: Neural network model acting as the prior for the restoration task.
        f_input_shape: Shape of the inputs to the model `f` (excluding batch size).
        z_std: Standard deviation for generating random input `z` (default 0.1).
        loss_mask: Binary mask (same shape as x0) to apply to images during loss calculation. Useful for inpainting.
        num_iters: Number of training iterations (default 1).
        init_lr: Initial learning rate for the Adam optimizer (default 0.01).
        save_filepath: Path to save the best model weights. If None, weights are not saved.

    Returns:
        Restored image after applying the DIP workflow.
    """
    # Generate a random input `z` for the neural network `f`
    shape = (1,) + f_input_shape  # Include batch dimension
    z = tf.constant(np.random.uniform(size=shape).astype(np.float32) * z_std, name='net_input')

    # Define the loss function
    def loss_fn(x_true, x):
        del x_true  # Ground-truth is not used directly in the loss calculation
        nonlocal x0, loss_mask
        if loss_mask is None:
            return tf.keras.losses.MSE(x, x0)  # Compute MSE between generated and input images
        else:
            return tf.keras.losses.MSE(x * loss_mask, x0 * loss_mask)  # Apply mask before loss calculation

    # Define metrics for evaluation
    def mse_to_gt(x_true, x):
        """Mean Squared Error between the generated image and ground-truth."""
        return tf.reduce_mean(tf.square(x - x_true))

    def psnr_to_gt(x_true, x, maxv=1.0):
        """Peak Signal-to-Noise Ratio (PSNR) between generated and ground-truth images."""
        mse = tf.reduce_mean(tf.square(x - x_true))
        psnr_ = 10.0 * tf.math.log(maxv**2 / mse) / tf.math.log(10.0)
        return psnr_

    # Compile the model with the optimizer and loss function
    opt = tf.keras.optimizers.Adam(learning_rate=init_lr)
    f.compile(optimizer=opt, loss=loss_fn, metrics=[mse_to_gt, psnr_to_gt])

    # Prepare callbacks for saving the best model
    callbacks = ()
    if save_filepath is not None:
        callbacks = create_saving_callback(save_filepath)

    # Train the model using the random input `z`
    history = f.fit(z,
                    x_true[None, ...],  # Add batch dimension to ground-truth image
                    epochs=num_iters,
                    steps_per_epoch=1,
                    verbose=0,
                    callbacks=callbacks+(TqdmCallback(verbose=1),))

    # Generate the restored image using the trained model
    x = f.predict(z)[0]

    # Visualize training dynamics and results
    fig = plt.figure(figsize=(10, 12), constrained_layout=True)
    gs = fig.add_gridspec(3, 2)
    axes = [fig.add_subplot(gs[0, :]),  # Training dynamics
            fig.add_subplot(gs[1, 0]),  # Input image
            fig.add_subplot(gs[1, 1]),  # Ground-truth image
            fig.add_subplot(gs[2, 0]),  # Last model output
            fig.add_subplot(gs[2, 1])]  # Best model output (optional)
    for ax in axes[1:]:
        ax.set_axis_off()  # Remove axes for images

    # Plot training history
    for key in history.history.keys():
        axes[0].plot(range(num_iters), history.history[key], label=key)
    axes[0].set_yscale('log')  # Log scale for better visualization
    axes[0].legend()
    axes[0].set_title("Training dynamics")

    # Display images
    axes[1].imshow(x0); axes[1].set_title('Input image')
    axes[2].imshow(x_true); axes[2].set_title('Ground-truth')
    axes[3].imshow(x); axes[3].set_title(f'Last output (PSNR = {psnr_to_gt(x_true, x):.2f})')

    # If the best model was saved, load it and display its output
    if save_filepath is not None and os.path.exists(save_filepath):
        f.load_weights(save_filepath)
        x_opt = f.predict(z)[0]
        axes[4].imshow(x_opt)
        axes[4].set_title(f'Best model output (PSNR = {psnr_to_gt(x_true, x_opt):.2f})')
    plt.show()

    # Return the restored image
    return x
```

```python
def create_saving_callback(filepath):
    """
    Creates a callback to save the best model during training.

    Args:
        filepath (str): The base file path where the model checkpoint will be saved.
                        The function appends the '.keras' extension to the file.

    Returns:
        tuple: A tuple containing a single ModelCheckpoint callback. This callback
               saves the model weights to the specified filepath whenever the monitored
               metric ('loss') improves during training.
    """
    return (
        tf.keras.callbacks.ModelCheckpoint(
            filepath=f"{filepath}.keras",  # Filepath with '.keras' extension
            monitor='loss',  # Monitors the 'loss' metric during training
            verbose=0,  # No detailed logs are printed when saving
            save_best_only=True,  # Saves the model only if the current 'loss' is the lowest
            mode='min'  # Looks for the minimum value of the 'loss' metric
        ),
    )


def deep_image_prior(input_shape,
                     noise_reg=None,
                     layers=(128, 128, 128, 128, 128),
                     kernel_size_down=3,
                     kernel_size_up=3,
                     skip=(0, 4, 4, 4, 4)):
    """
    Constructs a Deep Image Prior model using ImpulseNoiseWithDecay.

    Args:
        input_shape (tuple): Shape of the input image (height, width, channels).
        noise_reg (dict, optional): Parameters for Impulse noise regularization
                                    (e.g., {'prob': 0.1, 'decayrate': 0.9999, 'decaysteps': 100}).
        layers (tuple): Number of filters for each layer in the downsampling and upsampling stages.
        kernel_size_down (int): Kernel size for downsampling convolutional layers.
        kernel_size_up (int): Kernel size for upsampling convolutional layers.
        skip (tuple): Number of skip connection filters for each depth.

    Returns:
        tf.keras.Model: Compiled Deep Image Prior model.
    """

    def norm_and_active(x):
        """
        Applies batch normalization and LeakyReLU activation to the input tensor.
        """
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.LeakyReLU()(x)
        return x

    # Initialize model and input layer
    model = tf.keras.models.Sequential(name="Deep Image Prior")
    inputs = tf.keras.Input(shape=input_shape)

    ## Inputs
    x = inputs
    if noise_reg is not None:  # Add Impulse noise regularization if specified
        x = ImpulseNoiseWithDecay(**noise_reg)(x)

    ## Downsampling layers
    down_layers = []  # Store skip connections for later use
    for i, (num_filters, do_skip) in enumerate(zip(layers, skip)):
        # Add skip connections if specified
        if do_skip > 0:
            down_layers.append(
                norm_and_active(
                    tf.keras.layers.Conv2D(
                        filters=do_skip, kernel_size=1, strides=1, name=f"conv_skip_depth_{i}"
                    )(x)
                )
            )
        # Add downsampling convolutional layers
        for j, strides in enumerate([2, 1]):  # Strides=2 for downsampling, Strides=1 for refinement
            x = norm_and_active(
                tf.keras.layers.Conv2D(
                    num_filters, kernel_size_down, strides=strides, padding='same',
                    name=f"conv_down_{j + 1}_depth_{i}"
                )(x)
            )

    ## Upsampling layers
    for i, (num_filters, do_skip) in enumerate(zip(layers[::-1], skip[::-1])):
        # Upsample to the previous resolution
        x = tf.keras.layers.UpSampling2D(interpolation='bilinear', name=f"upsample_depth_{i}")(x)
        if do_skip:  # Concatenate skip connections if present
            x = tf.keras.layers.Concatenate(axis=-1)([x, down_layers.pop()])
        # Add upsampling convolutional layers
        for j, kernel_size in enumerate([kernel_size_up, 1]):
            x = norm_and_active(
                tf.keras.layers.Conv2D(
                    num_filters, kernel_size, strides=1, padding='same',
                    name=f"conv_up_{j + 1}_depth_{i}"
                )(x)
            )

    ## Output layer
    x = tf.keras.layers.Conv2D(filters=3, kernel_size=1, strides=1, name="conv_out")(x)
    x = tf.keras.layers.Activation('sigmoid')(x)  # Sigmoid activation to normalize outputs to [0, 1]
    return tf.keras.Model(inputs=inputs, outputs=x, name="deep_image_prior")
```

```python
import tensorflow as tf
import numpy as np

import tensorflow as tf
import numpy as np

class ImpulseNoiseWithDecay(tf.keras.layers.Layer):
    def __init__(self, prob, decayrate=0.99999, decaysteps=1, **kwargs):
        """
        A custom layer that adds impulse (salt-and-pepper) noise with decaying probability.

        Args:
            prob (float): Initial probability of impulse noise.
            decayrate (float): Decay rate for the noise probability. Multiplied
                               with the current probability after every decaystep.
            decaysteps (int): Number of calls after which the probability decays.
                              Default is 1.
            kwargs: Additional keyword arguments passed to the parent class.
        """
        super(ImpulseNoiseWithDecay, self).__init__(**kwargs)
        self.prob = prob
        self.decayrate = decayrate
        self.decaysteps = decaysteps
        self.num_calls = 0  # Tracks the number of times the layer is called

    def call(self, inputs, training=None):
        """
        Applies impulse noise to the inputs with decaying probability.

        Args:
            inputs (Tensor): Input tensor to the layer.
            training (bool): If True, noise is added. Otherwise, inputs are returned unchanged.

        Returns:
            Tensor: Input tensor with added impulse noise during training, or unchanged inputs during inference.
        """
        if training:
            # Increment the number of calls only during training
            self.num_calls += 1

            # Calculate the decayed probability
            prob = self.prob * self.decayrate ** (self.num_calls // self.decaysteps)

            # Generate random matrix to determine pixel alteration
            random_matrix = tf.random.uniform(tf.shape(inputs)[:2], minval=0, maxval=1)

            # Define salt_mask and pepper_mask here
            salt_mask = tf.ones_like(inputs) * 255.0 # White pixels
            pepper_mask = tf.zeros_like(inputs)      # Black pixels


            # Apply impulse noise to the image
            noisy_image = tf.where(random_matrix[:, :, None, None] < prob / 2, salt_mask, inputs)
            noisy_image = tf.where(random_matrix[:, :, None, None] > 1 - prob / 2, pepper_mask, noisy_image)

            return noisy_image
        else:
            # During inference, return inputs unchanged
            return inputs

def display_dip_model(input_shape=(256, 256, 3)):
    """
    Display the architecture summary of the Deep Image Prior model.

    Args:
        input_shape (tuple): The shape of the input image (default: (256, 256, 3)).
    """
    # Create and build the model with the specified input shape
    model = deep_image_prior(input_shape)
    model.build(input_shape)

    # Print the model summary
    print(model.summary())


x_true = resize(imread(imgrachgia_path).astype(np.float32) / 255., (256, 256))
x0 = add_impulse_noise(x_true, prob=0.3)
_, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(x0); axes[0].set_axis_off()
axes[0].set_title("Input noisy image")
axes[1].imshow(x_true); axes[1].set_axis_off()
axes[1].set_title("Ground-truth image")

plt.show()

input_shape = x0.shape
impulse_noise_reg = {'prob': 0.3, 'decayrate': 0.9999, 'decaysteps': 100}

model = deep_image_prior(input_shape, noise_reg=impulse_noise_reg)
x = dip_workflow(x0, x_true, model, input_shape, num_iters=3000, save_filepath='best_dip_denoising.hdf5')
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG19

def vgg19_denoising(input_shape=(256, 256, 3), noise_prob=0.3):
    """
    Constructs a VGG19-based denoising model.

    Args:
        input_shape (tuple): Shape of the input image (default: (256, 256, 3)).
        noise_prob (float): Probability of adding impulse noise to the image.

    Returns:
        tf.keras.Model: The VGG19-based model for denoising.
    """
    # Load VGG19 model without top (fully connected layers) and without weights
    base_model = VGG19(weights=None, include_top=False, input_shape=input_shape)

    # Freeze VGG layers to use as feature extractor
    base_model.trainable = False

    # Add custom layers on top
    x = base_model.output
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dense(256, activation='relu')(x)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    x = tf.keras.layers.Dense(np.prod(input_shape), activation='sigmoid')(x)
    x = tf.keras.layers.Reshape(input_shape)(x)

    # Create and compile the model
    model = tf.keras.Model(inputs=base_model.input, outputs=x)
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

    return model

def add_impulse_noise(image, prob=0.3):
    """
    Adds impulse (salt-and-pepper) noise to an image.

    Args:
        image (np.array): The input image.
        prob (float): Probability of adding noise to each pixel.

    Returns:
        np.array: Noisy image.
    """
    noisy_image = image.copy()
    salt_pepper = np.random.rand(*image.shape)
    noisy_image[salt_pepper < prob / 2] = 0  # Add pepper (black)
    noisy_image[salt_pepper > 1 - prob / 2] = 1  # Add salt (white)
    return noisy_image

def display_results(x0, x_true, x_pred):
    """
    Display the noisy image, ground truth, and denoised image.
    """
    _, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(x0)
    axes[0].set_title("Noisy Image")
    axes[1].imshow(x_true)
    axes[1].set_title("Ground Truth")
    axes[2].imshow(x_pred)
    axes[2].set_title("Denoised Image")
    plt.show()

# Example usage
x_true = resize(imread(imgrachgia_path).astype(np.float32) / 255., (256, 256))
x0 = add_impulse_noise(x_true, prob=0.3)

# Prepare model
model = vgg19_denoising(input_shape=(256, 256, 3))

# Train the model
model.fit(np.expand_dims(x0, axis=0), np.expand_dims(x_true, axis=0), epochs=100, batch_size=1)

# Predict the denoised image
x_pred = model.predict(np.expand_dims(x0, axis=0))[0]

# Display results
display_results(x0, x_true, x_pred)
```

```python
def psnr(x_true, x_pred, maxv=1.0):
    """Calculate Peak Signal-to-Noise Ratio."""
    mse = np.mean((x_true - x_pred) ** 2)
    return 10 * np.log10(maxv**2 / mse)

def evaluate_model(x_true, x_pred):
    """Evaluate the model using PSNR and MSE."""
    mse_value = np.mean((x_true - x_pred) ** 2)
    psnr_value = psnr(x_true, x_pred)
    return mse_value, psnr_value

# Evaluate the model
mse_value, psnr_value = evaluate_model(x_true, x_pred)
print(f'MSE: {mse_value:.4f}, PSNR: {psnr_value:.4f}')
```
